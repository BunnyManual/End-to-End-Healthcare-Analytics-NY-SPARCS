## End-to-End Healthcare Analytics: NY SPARCS ##

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df_los = pd.read_csv('los_distribution.csv')
    plt.figure(figsize=(15, 8))
    sns.set_style("whitegrid")
    sns.barplot(x='Length of Stay', y='patient_count', data=df_los, color='skyblue')
    plt.title('Distribution of Hospital Length of Stay (1-30 Days)', fontsize=16)
    plt.xlabel('Length of Stay (Days)', fontsize=12)
    plt.ylabel('Number of Patients', fontsize=12)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig('los_distribution_plot.png')
    print("Successfully generated 'los_distribution_plot.png' on your computer.")
except FileNotFoundError:
    print("This script could not find 'los_distribution.csv' in the same folder.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
try:
    df_diag = pd.read_csv('top_diagnoses.csv')
    plt.figure(figsize=(12, 10))
    sns.set_style("whitegrid")
    barplot = sns.barplot(
        x='number_of_cases',
        y='diagnosis',
        data=df_diag,
        palette='viridis',
        orient='h'
    )
    plt.title('Top 15 Most Common Diagnoses for Inpatient Discharges', fontsize=16)
    plt.xlabel('Number of Cases', fontsize=12)
    plt.ylabel('Diagnosis', fontsize=12)
    plt.tight_layout()
    plt.savefig('top_diagnoses_plot.png')
    print("Successfully generated 'top_diagnoses_plot.png' on your computer.")
except FileNotFoundError:
    print("This script could not find 'top_diagnoses.csv' in the same folder.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
try:
    df_sev = pd.read_csv('severity_vs_charges.csv')
    plt.figure(figsize=(10, 7))
    sns.set_style("whitegrid")
    bar_plot = sns.barplot(
        x='severity_level',
        y='average_charge',
        data=df_sev,
        palette='plasma'
    )
    plt.title('Average Hospital Charges by Patient Illness Severity', fontsize=16)
    plt.xlabel('Severity of Illness', fontsize=12)
    plt.ylabel('Average Total Charges ($)', fontsize=12)
    from matplotlib.ticker import FuncFormatter
    formatter = FuncFormatter(lambda y, _: f'${int(y):,}')
    plt.gca().yaxis.set_major_formatter(formatter)
    plt.tight_layout()
    plt.savefig('severity_vs_charges_barchart.png')
    print("Successfully generated 'severity_vs_charges_barchart.png'")
    plt.figure(figsize=(10, 8))
    plt.pie(
        df_sev['number_of_cases'],
        labels=df_sev['severity_level'],
        autopct='%1.1f%%',
        startangle=140,
        colors=sns.color_palette('plasma', n_colors=len(df_sev))
    )
    plt.title('Distribution of Patients by Illness Severity', fontsize=16)
    plt.ylabel('')
    plt.tight_layout()
    plt.savefig('severity_patient_distribution_piechart.png')
    print("Successfully generated 'severity_patient_distribution_piechart.png'")
except FileNotFoundError:
    print("This script could not find 'severity_vs_charges.csv' in the same folder.")
except Exception as e:
    print(f"An error occurred: {e}")
```
